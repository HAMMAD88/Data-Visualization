```python
from tables.unimplemented import Unknown
#mercator
from vega_datasets import data
import altair as alt
import pandas as pd
data1   = pd.read_csv('2020_2.csv')
# countries = alt.topo_feature(data.world_110m.url, 'countries')
# https://github.com/deldersveld/topojson/blob/master/countries/united-kingdom/uk-counties.jso
UK = alt.topo_feature('https://raw.githubusercontent.com/deldersveld/topojson/master/continents/europe.json', 'continent_Europe_subunits')
background = alt.Chart(UK).mark_geoshape(
    fill='lightgray',
    stroke='grey',
).encode(tooltip=['properties.geounit:N']).project('mercator').properties(
    width=200,
    height=200,
    title='Europe (Mercator)'
)
points = alt.Chart(data1).mark_circle().encode(
     longitude='Longitude:Q',
     latitude='Latitude:Q',
    #  size=alt.value(25),
     tooltip = ['Name','Departures:Q'],
     size = 'Departures:Q',
     color = alt.Color('Departures:Q',scale=alt.Scale(scheme='orangered')),
)
background + points
```
